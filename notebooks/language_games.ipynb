---
title: "Language Games"
layout: page
---

I will argue that one can consider LLMs to be playing language games, in the sense of Wittgenstein. In particular, the method of Reinforcement Learning from Human Feedback (RLHF) is essentially teaching the model to play a game. Therefore, if the reductive view of Wittgenstein with regards to language is correct, one can consider LLMs to be genuinely capable of language. Indeed, if that view is accepted, there is little difference between the way LLMs communicate and the way we do.
